## missing values

```python
df.isnull()
```

```python
df.isnull().sum() #col wise null count
```

```python
#few nulls as comapared to 25731.so, we can fill nulls
```

```python
df.dtypes
```

```python
#filling int/float cols with median..,

num_col=df.select_dtypes(include=['int','float']).columns
df[num_col]=df[num_col].fillna(df[num_col].median())

```

```python
#obj col nulls -> false
obj_col=['resoFacts/hasHeating','resoFacts/hasCooling','resoFacts/hasFireplace','resoFacts/isNewConstruction']
df[obj_col]=df[obj_col].fillna(False)
```

```python
df[obj_col]=df[obj_col].replace({True:1,False:0})
```

```python
df.isnull().sum() #nulls filled
```

## changing datatype

```python
df.dtypes
```

```python
#obj,bool to int
df[obj_col]=df[obj_col].astype(int)
df['resoFacts/hasGarage']=df['resoFacts/hasGarage'].astype(int)
```
